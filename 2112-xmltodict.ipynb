```python
import xmltodict
import json

xml = """
<mydocument has="an attribute">
  <and>
    <many>elements</many>
    <many>more elements</many>
  </and>
  <plus a="complex">
    element as well
  </plus>
</mydocument>
"""

print(json.dumps(xmltodict.parse(xml), indent=4))
```

```python
dict = xmltodict.parse(xml)
dict['mydocument']
```

```python
dict['mydocument']['and']['many']
```

```python
xmlFile = "data/sample.bpmn"
with open(xmlFile) as fobj:
    xml2 = fobj.read()
    
dict2 = xmltodict.parse(xml2)
print(json.dumps(dict2, indent=2))
```
